```python
import numpy as np
data = np.load("meuwly_db_HCOOH_MP2.npz")   
```

```python
energies = data["E"]
energies.shape
```

```python
indices = np.random.permutation(data['E'].shape[0])
print(data.files)
print(indices)
training_idx, test_idx = indices[:500], indices[500:]
print(type(training_idx))
print(training_idx.shape)
print(test_idx.shape)
```

```python
numAtoms=data["N"]
# print(f"Number of Atoms:\n {numberAtoms}")
print("Number of Atoms Shape: ", numAtoms.shape)
numAtoms_training = numAtoms[training_idx]
numAtoms_test = numAtoms[test_idx]
print(numAtoms_training.shape)

geo = data["R"]
# print(f"Geometries:\n {geo}")
print("Geometry Shape: ", geo.shape)
geo_training = geo[training_idx]
geo_test = geo[test_idx]

totalCharge=data["Q"]
# print(f"Total charge:\n {totalCharge}")
print("Total charge shape: ", totalCharge.shape)
totalCharge_training = totalCharge[training_idx]
totalCharge_test = totalCharge[test_idx]

dipoleMoment=data["D"]
# print(f"Dipole moment:\n {dipoleMoment}")
print("Dipole moment shape: ", dipoleMoment.shape)
dipoleMoment_training = dipoleMoment[training_idx]
dipoleMoment_test = dipoleMoment[test_idx]

energies = data["E"]
# print(f"Energies:\n {energies}")
print("Energy Shape: ", energies.shape)
energies_training = energies[training_idx]
energies_test = energies[test_idx]

grad=data["F"]
# print(f"Gradient:\n {grad}")
print("Gradient Shape: ", grad.shape)
grad_training = grad[training_idx]
grad_test = grad[test_idx]

atomicNum=data["Z"]
# print(f"Atomic number:\n {atomicNumber}")
print("Atomic number shape: ", atomicNum.shape)
atomicNum_training = atomicNum[training_idx]
atomicNum_test = atomicNum[test_idx]
```

```python
np.savez('meuwly_db_HCOOH_MP2_500.npz', N=numAtoms_training, R=geo_training, Q=totalCharge_training, D=dipoleMoment_training, E=energies_training, F=grad_training, Z=atomicNum_training)
# np.savez('HCOOH_ccf12_Test_rmNoises.npz', N=numAtoms_test, R=geo_test, Q=totalCharge_test, D=dipoleMoment_test, E=energies_test, F=grad_test, Z=atomicNum_test)
```
